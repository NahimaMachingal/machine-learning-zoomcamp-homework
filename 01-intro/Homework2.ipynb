```python
for seed in seeds:
    # Split dataset: 60% train, 20% validation, 20% test
    train_val_df, test_df = train_test_split(df, test_size=0.2, random_state=seed)
    train_df, val_df = train_test_split(train_val_df, test_size=0.25, random_state=seed)  # 0.25*0.8 = 0.2
    
    X_train = train_df.drop('fuel_efficiency_mpg', axis=1)
    y_train = train_df['fuel_efficiency_mpg']
    
    X_val = val_df.drop('fuel_efficiency_mpg', axis=1)
    y_val = val_df['fuel_efficiency_mpg']
    
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    preds = model.predict(X_val)
    rmse = mean_squared_error(y_val, preds) ** 0.5
    rmse_scores.append(rmse)
```

```python
std_rmse = np.std(rmse_scores)
```
